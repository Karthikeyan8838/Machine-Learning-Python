```python
data=data.drop(['duration','campaign'],axis=1)
data.head()
```

```python
data.drop(['month','day_of_week'],axis=1,inplace=True)
data.head()
```

```python
data['y']=data['y'].replace(['no','yes'],[0,1])
data.head()
```

```python
data=pd.get_dummies(data)
data.head()
```

```python
x=data.drop(['y'],axis=1)
y=data['y']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=123)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
x_test
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
model.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("ACCURACY SCORE OF LOGISTIC REGRESSION: ",accuracy_score(y_test,y_pred))
print("CONFUSION MATRIX OF LOGISTIC REGRESSION: ",confusion_matrix(y_test,y_pred),sep='\n')
print("CLASSIFICATION REPORT OF LOGISTIC REGRESSION: ",classification_report(y_test,y_pred),sep='\n')
```

```python
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score,auc
from sklearn.metrics import roc_curve
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=123)
```

```python
c=GaussianNB()
c.fit(X_train,Y_train)

Y_pred=c.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("ACCURACY SCORE OF NAIVE BAYES: ",accuracy_score(Y_test,Y_pred))
print("CONFUSION MATRIX OF NAIVE BAYES: ",confusion_matrix(Y_test,Y_pred),sep='\n')
print("CLASSIFICATION REPORT OF NAIVE BAYES: ",classification_report(Y_test,Y_pred),sep='\n')
```

```python
##KNN
from sklearn.neighbors import KNeighborsClassifier
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.25,random_state=0)
e=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
e.fit(xtrain,ytrain)
ypred=e.predict(xtest)
e.predict_proba(xtest)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("ACCURACY SCORE OF NAIVE BAYES: ",accuracy_score(ytest,ypred))
print("CONFUSION MATRIX OF NAIVE BAYES: ",confusion_matrix(ytest,ypred),sep='\n')
print("CLASSIFICATION REPORT OF NAIVE BAYES: ",classification_report(ytest,ypred),sep='\n')
```

```python
from sklearn.svm import SVC
k=SVC(probability=True)
x__train,x__test,y__train,y__test=train_test_split(x,y,test_size=.20,random_state=0)
k.fit(x__train,y__train)
y__pred=k.predict(x__test)
k.predict_proba(x__test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print("ACCURACY SCORE OF NAIVE BAYES: ",accuracy_score(y__test,y__pred))
print("CONFUSION MATRIX OF NAIVE BAYES: ",confusion_matrix(y__test,y__pred),sep='\n')
print("CLASSIFICATION REPORT OF NAIVE BAYES: ",classification_report(y__test,y__pred),sep='\n')
```
